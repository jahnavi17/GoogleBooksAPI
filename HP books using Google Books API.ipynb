# Getting all Harry Potter books available on Google using Google Books API

In this project, I used Google Books API along with Authorization with API key. I got the personalized API key using Google console. I got all the books that has "Harry Potter" in the title along with book_id,author,No of pages and average rating.

```python
import requests
import pandas as pd
```

```python
requests.get("https://www.googleapis.com/books/v1/volumes?q=intitle:Harry+Potter&key=YourAPIKey").json()
```

```python
ids=[]
titles=[]
authors=[]
pages=[]
ratings=[]
for i in range(10):
    x=str(i*40)
    req=requests.get("https://www.googleapis.com/books/v1/volumes?q=intitle:Harry+Potter&startIndex="+x+"&maxResults=40&key=YourAPIKey")
    json1=req.json()
    if ('items') in json1.keys():
        for j in range(len(json1['items'])):
            ids.append(json1['items'][j]['id'])
            titles.append(json1['items'][j]['volumeInfo']['title'])
            if ('authors') in (json1['items'][j]['volumeInfo']).keys():
                authors.append(json1['items'][j]['volumeInfo']['authors'])
            else:
                authors.append("NA")
            if ('pageCount') in (json1['items'][j]['volumeInfo']).keys():
                pages.append(json1['items'][j]['volumeInfo']['pageCount'])
            else:
                pages.append("NA")
            if ('averageRating') in (json1['items'][j]['volumeInfo']).keys():
                ratings.append(json1['items'][j]['volumeInfo']['averageRating'])
            else:
                ratings.append("NA")
```

```python
print(len(ids))
print(len(titles))
print(len(authors))
print(len(pages))
print(len(ratings))
```

```python
df=pd.DataFrame({'book_id':ids,'book_title':titles,'author':authors,'no_on_pages':pages,'average_rating':ratings})
```

```python
print(df)
```

```python
data=df[df['book_title'].str.contains("Harry Potter")]
```
